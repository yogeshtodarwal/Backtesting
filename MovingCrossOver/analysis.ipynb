```python
import pandas as pd

# Load the CSV file
file_path = 'all_trades_detailed.csv'
data = pd.read_csv(file_path)

# Filter out the relevant columns
filtered_data = data[['Fast Period', 'Slow Period', 'profit']]

# Group by Fast Period and Slow Period and calculate the statistics
grouped_stats = filtered_data.groupby(['Fast Period', 'Slow Period']).agg(
    total_profit=('profit', 'sum'),
    mean_profit=('profit', 'mean'),
    median_profit=('profit', 'median'),
    min_profit=('profit', 'min'),
    max_profit=('profit', 'max'),
    total_trades=('profit', 'count'),
    profitable_trades=('profit', lambda x: (x > 0).sum())
).reset_index()

# Calculate the percentage of profitable trades
grouped_stats['percentage_profitable'] = (grouped_stats['profitable_trades'] / grouped_stats['total_trades']) * 100


# Display the grouped statistics
grouped_stats
```

```python

# Set the maximum number of rows to display
pd.set_option('display.max_rows', 81)

# Sort the dataframe based on percentage_profitable column in descending order
grouped_stats_sorted = grouped_stats.sort_values('percentage_profitable', ascending=False)

# Print all the rows and columns of grouped_stats_sorted
grouped_stats_sorted
```

```python
import pandas as pd

# Assume 'data' is your DataFrame and it contains a column 'profit'

# Calculate the first quartile (Q1)
Q1 = data['profit'].quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = data['profit'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Calculate the lower bound
lower_bound = Q1 - 1.5 * IQR

# Calculate the upper bound
upper_bound = Q3 + 1.5 * IQR

# Detect the outliers
outliers = data[(data['profit'] < lower_bound) | (data['profit'] > upper_bound)]

# Remove the outliers
data_no_outliers = data[(data['profit'] >= lower_bound) & (data['profit'] <= upper_bound)]

# Calculate the mean profit without outliers
mean_profit_no_outliers = data_no_outliers['profit'].mean()

# Calculate the mean profit with outliers
mean_profit_with_outliers = data['profit'].mean()

# Calculate the median profit without outliers
median_profit_no_outliers = data_no_outliers['profit'].median()

# Calculate the median profit with outliers
median_profit_with_outliers = data['profit'].median()

# Calculate the standard deviation of profit without outliers
std_dev_profit_no_outliers = data_no_outliers['profit'].std()

# Calculate the standard deviation of profit with outliers
std_dev_profit_with_outliers = data['profit'].std()

# Create a DataFrame to display the results
results = pd.DataFrame({
    'Metric': ['Mean Profit', 'Median Profit', 'Standard Deviation of Profit'],
    'With Outliers': [mean_profit_with_outliers, median_profit_with_outliers, std_dev_profit_with_outliers],
    'Without Outliers': [mean_profit_no_outliers, median_profit_no_outliers, std_dev_profit_no_outliers]
})

# Display the results
print(results)
```
